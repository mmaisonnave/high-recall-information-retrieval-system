```python
# ~ IMPORTS ~
import numpy as np
import sys
sys.path.append('/home/ec2-user/SageMaker/mariano/repositories/tdmstudio-high-recall-information-retrieval-system/')
from utils.io import info
from utils.tdmstudio import get_title, get_filename
import pandas as pd
import os

# ~ INPUTS ~
filename='/home/ec2-user/SageMaker/serperi/system/sessions/scal/One_second_round/data/exported_data_2022-11-23_00-18.csv'

# ~ OUTPUTS ~
evaluation_file= os.path.join('/home/ec2-user/SageMaker/serperi/system/sessions/scal/One_second_round/data/',
                              'exported_data_for_evaluation_2022-11-23_00-18.csv')


# ~ READING Suggestions from file ~
df = pd.read_csv(filename)
info(f'df.shape={df.shape} (before removing labeled as relevant)')
df = df[df['relevant_or_suggested']=='sugg']
info(f'df.shape={df.shape} (only suggestions)')
df = df.sort_values(by=['confidence'], ascending=False)
df['index']=range(df.shape[0])
df['id']=list(map(lambda url: url.split('/')[-1], df['URL']))
df.head()

# ~ DIVIDING in ranges and taking 50 random samples ~
rng = np.random.default_rng(1018)
ranges = [(0, 6), (6, 36), (36, 221), (221, 1341), (1341, 8115)]
count_per_range = [2, 2, 6, 12, 28] 

selected = pd.DataFrame({}, columns=df.columns)
for count, (ini, fin) in zip(count_per_range, ranges):
    selected_idx = rng.choice(range(ini,fin), size=count )
    assert all([ini<=arg and arg < fin for arg in selected_idx])
    selected = selected.append([df.iloc[selected_idx,:]])
    info(f'From range [{ini:4}, {fin:4}) selecting {count:2} instances (df.shape={selected.shape}).')
info(f'selected.shape={selected.shape}')

selected['range'] = [(r+1) for idx,count in enumerate(count_per_range) for r in [idx]*count]
# selected['range'] = [ for list_ in 4]

# ~ ADDING titles ~ 
selected['file']=list(map(get_filename, selected['id']))
selected['title']=list(map(get_title, selected['file']))
selected=selected.drop(columns=['file','id', 'relevant_or_suggested', 'confidence', 'index'])
selected

# ~ REORDERING columns ~

selected = selected[['range','URL','title']]

# ~ RANDOMIZING and DUMPING to disk ~
new_order = list(range(selected.shape[0]))
rng.shuffle(new_order)

selected = selected.iloc[new_order,]
selected.to_csv(evaluation_file, index=False)
info(f'Exporting to: {evaluation_file}')


selected
```
