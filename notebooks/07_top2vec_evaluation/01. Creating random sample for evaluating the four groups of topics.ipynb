```python
import pandas as pd
import numpy as np
import os
import sys
sys.path.append('/home/ec2-user/SageMaker/mariano/repositories/tdmstudio-high-recall-information-retrieval-system/')

from utils import io
```

```python
# CONSEGUIR INFORMACIÃ“N DE RELEVANCIA DE TÃ“PICO DEL SPREADSHEET
df = pd.read_csv(topics_file)
df.head()
```

```python

topic_relevance_df = pd.read_csv(topics_relevance_file, sep=';')

print(f"Number of relevant topics:                "\
      f" {np.sum(topic_relevance_df['topic relevance']=='R')}")
print(f"Number of irrelevant topics:               "\
      f"{np.sum(topic_relevance_df['topic relevance']=='IR')}")
print(f"Number of irrelevant topics (with rel):    "\
      f"{np.sum((topic_relevance_df['topic relevance']=='IR') & (topic_relevance_df['with_relevant']==True))}")
print(f"Number of irrelevant topics (without rel): "\
      f"{np.sum((topic_relevance_df['topic relevance']=='IR') & (topic_relevance_df['with_relevant']==False))}")

topic_relevance_df['has_labeled_as_relevant'] = [type(elem)!=float for elem in df['relevants'] ]
topic_relevance_df.to_csv('tmp.txt')
topic_relevance_df.head()
```

```python
relevant_topic_mask=topic_relevance_df['topic relevance']=='R'
irrelevant_topic_mask = topic_relevance_df['topic relevance']=='IR'
prone_to_be_relevant_mask=topic_relevance_df['topic relevance']=='PRONE TO BE RELEVANT'

has_relevant_mask=(topic_relevance_df['with_relevant']==True)|(topic_relevance_df['has_labeled_as_relevant'])


masks = [relevant_topic_mask,
         (irrelevant_topic_mask) & (has_relevant_mask),
         (irrelevant_topic_mask) & (~has_relevant_mask),
         prone_to_be_relevant_mask,
        ]
group_data = {'group':[1,2,3,4],
             'group description': ['RELEVANT', 'IRRELEVANT (with rel)', 'IRRELEVANT (without rel)', 'PRONE TO REL'],
             'topic ids': [[], [], [], []],
              'suggestion ids':[[], [], [], []],
              'topic count':[0, 0, 0, 0],
              'suggestion count':[0, 0, 0, 0],
              'relevant count':[0, 0, 0, 0],
              'relevant ids':[[], [], [], []],
             }

for group_ix in range(4):
    auxdf = topic_relevance_df[masks[group_ix]]
    print(f'group {group_ix} - {auxdf.shape}')
    group_data['topic ids'][group_ix]=list(auxdf['topic no.'])
    group_data['topic count'][group_ix] =len(list(auxdf['topic no.']))
    
    
for group_ix in range(4):
    for topic_id in group_data['topic ids'][group_ix]:
        suggestions = df[ df['topic no.'] == topic_id]['suggestions'].iloc[0].split(';')

        
        if type(df[df['topic no.']==topic_id]['relevants'].iloc[0])==str:
            relevants=(df[df['topic no.']==topic_id]['relevants'].iloc[0].split(';'))
            group_data['relevant ids'][group_ix]+=(relevants)
            group_data['relevant count'][group_ix]+=len(relevants)

            
        group_data['suggestion ids'][group_ix] += suggestions
        group_data['suggestion count'][group_ix] += len(suggestions)
        
group_df = pd.DataFrame(group_data)

group_df
```

# RANDOM SAMPLING

```python
# np.random.randint(65535)
# 23173
seed=23173
sample_size=25
group_count=4

io.info(f'SAMPLE SIZE={sample_size}')
io.info(f'OUTPUT FILE={evaluation_file}')
if os.path.isfile(evaluation_file):
    io.warning('OUTPUT FILE exists, overwriting ...')
rng = np.random.default_rng(seed)

relevant_ids = set()
for group_ix in range(4):
    relevant_ids=relevant_ids.union(group_df.iloc[group_ix,:]['relevant ids'])

io.info(f'Number of relevant IDs found: {len(relevant_ids)}')

data = {'group id': [],
        'group name': [],
        'id': [],
        'label': [],        
       }

relevant_sampled=0
for group_ix in range(group_count):
    auxdf = group_df.iloc[group_ix,:]
    documents = auxdf['suggestion ids'] + auxdf['relevant ids']
    sample = rng.choice(documents, size=sample_size, replace=False, )
    for elem in sample:
        data['group id'].append(group_ix)
        data['group name'].append(group_df['group description'].iloc[group_ix])
        data['id'].append(elem)
        data['label'].append('relevant' if elem in relevant_ids else 'unknown')
        if elem in relevant_ids:
            relevant_sampled+=1
            
io.info(f'Relevant found while sampling {relevant_sampled}')
outputdf = pd.DataFrame(data)
assert  len(set(outputdf['id']))==group_count*sample_size
outputdf.to_csv(evaluation_file)
outputdf
```
