```python
import sys
sys.path.append('/home/ec2-user/SageMaker/mariano/repositories/tdmstudio-high-recall-information-retrieval-system/')
from utils import tdmstudio

import pandas as pd

df = pd.read_csv(datapath)
df['text'] = list(map(lambda id_: tdmstudio.get_title_and_text(tdmstudio.get_filename(str(id_))), df['id']))
df.head()
```

```python
id2newscore={id_:df[df['id']==id_]['new_score'].iloc[0] for id_ in df['id']}
assert all([df[df['id']==id_]['new_score'].shape[0]==1 for id_ in df['id']])
```

```python
topic_df = pd.read_csv(topics_csv)
topic_df.head()
```

```python
import numpy as np

data = {'id':[], 'suggestions by relevance':[]}

topic2sortedsuggestions={}

for i in range(topic_df.shape[0]):
    suggestions = np.array(topic_df['suggestions'].iloc[i].split(';'))
    scores = np.array([id2newscore[int(id_)] for id_ in suggestions])
    suggestions = suggestions[np.argsort(scores)[::-1]]
    topic_no=topic_df['topic no.'].iloc[i]
    data['id'].append(topic_no)
    data['suggestions by relevance'].append(';'.join(suggestions))
    topic2sortedsuggestions[topic_no]=suggestions
    
output_df = pd.DataFrame(data)
output_df.to_csv(sorted_suggestions, index=False)
output_df.head()
```
