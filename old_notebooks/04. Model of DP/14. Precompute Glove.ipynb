```python
from utils.general import info,ok, warning
import os
from tqdm import tqdm
from numpy import save, load
import pickle

def combine_files(file_):
    id_ = file_[:-13]
    g300 = 'precomputed/'+id_+'_glove300.npy'
    g600 = 'precomputed/'+id_+'_glove600.npy'
    if os.path.isfile(g300) and os.path.isfile(g600):
        print('entro!')
        output = 'precomputed/'+id_+'_glove.p'


        data = [load(g300),load(g600)]
        assert data[0].shape==(300,) and data[1].shape==(600,)
        pickle.dump(data,open(output,'wb'))
        if os.path.isfile(output):
            print('exit when saving')
            os.remove(g300)
            os.remove(g600)
        else:
            warning('FAILED TO SAVE.')
        
```

```python
files = [f for f in os.listdir('precomputed/') if f.endswith('_glove600.npy')]
files
```

```python
%%time
from utils.general import info, ok
import concurrent.futures
import datetime
writer = open('done.txt', 'w')
writer.write(f'{datetime.datetime.now()} Starting...\n')

info('Starting...')

with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    executor.map(combine_files, files)
    
    
writer.write(f'{datetime.datetime.now()} Done!')
writer.close()
ok('Done!')
```

```python
import spacy
from numpy import load,save
from utils.tdmstudio import TDMStudio
nlp = spacy.load('en_core_web_lg', disable=['textcat','ner','parser','lemmatizer','tagger'])
nlp.tokenizer

def process_file(file_):
    title,text = TDMStudio.get_title_and_text(file_)
    if not title is None and not text is None:
        output = f"./precomputed/{file_.split('/')[-1][:-4]}_glove300.npy"
        if not os.path.isfile(output):
            X = nlp.tokenizer(f'{title}. {text}').vector
            save(output, X)
            del(X)
        del(output)
    del(title,text)
process_file(files[2])
```
